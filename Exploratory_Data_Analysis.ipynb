```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('/Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
# This will give top 5 records from data set.If we we want to get more then mention a number in head function.Ex: df.head(10)
df.head()
```

```python
# This will give a over all schema of table with datatypes
df.info()
```

```python
# remove unwanted colums from table.As we have 'Status' and 'unnamed1' as null values.So we are removing them
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
# Now we will check wether they have been removed or not
df.info()
```

```python
# Now we will check which coloumns have null values and thier count
pd.isnull(df).sum()
```

```python
# remove those null values
df.dropna(inplace=True)
```

```python
# Now we will check wether they have been removed or not
pd.isnull(df).sum()
```

```python
# Now we will change data type of a column
df['Amount'] = df['Amount'].astype('int')
```

```python
# Now we will check data type has been changed or not
df.info()
```

```python
# or you can use this method also
df['Amount'].dtypes
```

```python
# To find column names of dataset
df.columns
```

```python
# We will rename a column
df.rename(columns= {'Marital_Status':'Marriage_Pelli'},inplace=True)
```

```python
# Check the changed column name
df.columns
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df.describe()
```

```python
# use describe() for specific columns
df[['Age', 'Orders', 'Amount']].describe()
```

# BASED ON GENDER

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

FROM ABOVE GRAPH WE CAN CONCLUDE THERE ARE MORE FEMALE CUSTOMERS THAN MALE CUSTOMERS

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

FROM ABOVE GRAPH WE CAN CONCLUDE THERE ARE MORE FEMALE CUSTOMERS THAN MALE CUSTOMERS

# BASED ON AGES

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

# BASED ON STATE

```python
# total number of orders from top 7 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(7)

sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 7 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(7)

sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


#BASED ON MARRIAGE_PELLI

```python
ax = sns.countplot(data = df, x = 'Marriage_Pelli')

sns.set(rc={'figure.figsize':(5,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Divide male and female based on Marriage_Pelli
sales_state = df.groupby(['Marriage_Pelli', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marriage_Pelli',y= 'Amount', hue='Gender')
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

# BASED ON OCCUPATION

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

# BASED ON PRODUCT CATEGORY

```python
sns.set(rc={'figure.figsize':(30,8)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

#BASED ON PRODUCT_ID AND ORDERS

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
# top 10 most sold products (same thing as above)
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
